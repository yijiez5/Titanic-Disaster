```python
#data cleaning
```

```python
df_train=df_train.drop(['Cabin'],axis=1)
```

```python
df_train=df_train.drop(['Ticket'],axis=1)
```

```python
df_train.columns
```

```python
df_train['Title']=df_train.Name.str.extract(' ([A-Za-z]+)\.',expand=False)
pd.crosstab(df_train['Title'],df_train['Sex'])
```

```python
df_train['Title']=df_train['Title'].replace(['Capt','Col','Countess','Don','Dr','Jonkheer','Major','Master','Rev'],'Rare')
df_train['Title']=df_train['Title'].replace(['Mlle','Lady','Ms'],'Miss')
df_train['Title']=df_train['Title'].replace('Mme','Mrs')
df_train['Title']=df_train['Title'].replace('Sir','Mr')
df_train[['Title','Survived']].groupby(['Title'],as_index=False).mean()
```

```python
title_mapping={'Rare':1,'Miss':2,'Mrs':3,'Mr':4}
df_train['Title']=df_train['Title'].map(title_mapping)
df_train['Title']=df_train['Title'].fillna(0)
```

```python
age_title_mapping = {1: 'baby', 2: 'young adult', 3: 'adult', 4: 'young adult'}
```

```python
for x in range(len(df_train['AgeGroup'])):
    if df_train['AgeGroup'][x]=='unknown':
        df_train['AgeGroup'][x]=age_title_mapping[df_train['Title'][x]]
```

```python
age_mapping = {'baby': 1, 'child': 2, 'teenager': 3, 'young adult': 4, 'adult': 5, 'senior': 6}
df_train['AgeGroup'] = df_train['AgeGroup'].map(age_mapping)
```

```python
df_train=df_train.drop(['Age'],axis=1)
```

```python
df_train=df_train.drop(['Name'],axis=1)
```

```python
sex_mapping={'male':1,'female':0}
df_train['Sex']=df_train['Sex'].map(sex_mapping)
df_train.head()
```

```python
embarked_mapping={'S':0,'C':1,'Q':2}
df_train['Embarked']=df_train['Embarked'].map(embarked_mapping)
df_train.head()
```

```python
df_train['FareBand']=pd.qcut(df_train['Fare'],4,labels=[1,2,3,4])
df_train=df_train.drop(['Fare'],axis=1)
```
